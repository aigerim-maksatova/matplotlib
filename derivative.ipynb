# EXPONENTIAL GROWTH

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 7) 
y = 3 ** x

plt.figure(figsize = (8, 6))
plt.plot(x, y, label = "y = $3^x$")
plt.legend()
plt.grid(linewidth = 0.3)
plt.show()
```


**The curve shows a rapid increase**

```python
# its growth starts from points 2-3, so im going to plot derivative from point 3
```

# ITS DERIVATIVE:

```python
x = np.linspace(3, 8)
y = 3 ** x

derivative = 3**x * np.log(3)

plt.figure(figsize = (5, 5))
plt.plot(x, y, label="y = 3^x")
plt.plot(x, derivative, label = "derivative")
plt.legend()
plt.grid(linestyle = "--", linewidth = 0.5)
plt.show()
```

**it grows exponentially, just like the original function!!!**

# PARABOLA

```python
x = np.linspace(-10, 25)
y = x**2
plt.figure(figsize = (5, 5))
plt.plot(x, y)
plt.grid(linewidth = 0.3)
plt.show()
```

# DERIVATIVE of parabola:

```python
x = np.linspace(-10, 20)

derivative = 2*x

plt.figure(figsize = (6, 8))
plt.ylim(-20, 200)
plt.plot(x, y, label= "y = $x^2$")
plt.plot(x, derivative, label = "y' = 2x", linestyle = "--")

plt.title("$x^2$ and its Slope y' = 2x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(linewidth = 0.3)
plt.legend()
plt.show()
```

**The rate of change (the slope) of the function increases linearly**

**Means, there is constant rate of increase**

# SIN(X) AND ITS DERIVATIVE:

```python
from math import sin

dx = 0.1      # the change in x value
 
x = [x for x in np.arange(0,10,dx)]
y = [sin(x) for x in np.arange(0,10,dx)]
derivative = np.gradient(y, dx)

plt.plot(x, y, label = "sin(x)")
plt.plot(x, derivative, label = "derivative", linestyle='--')
plt.legend(loc = "lower right")
plt.grid(linewidth = 0.3)
plt.show()
```

**The slope is determined by the derivative of the function:**

**cos(x) fluctuates between -1 and 1, so it depends on (x)**
