```python
import matplotlib.pyplot as plt

plt.figure(figsize = (1, 1))
plt.axis('off')  

plt.text(0.5, 0.5, r'$e^{i\pi} + 1 = 0$', fontsize = 20, va = "center")

plt.show()
```

```python
import numpy as np

x = np.linspace(0, 2*np.pi, 200)
sin = np.cos(x)
cos = np.sin(x)

plt.figure(figsize=(6, 6))
plt.plot(sin, cos, label = "Circle: $e^{i\\theta}$")

# set the points
plt.scatter([1, -1], [0, 0])    
plt.scatter([0], [0], color='r')     

# Marking that points
plt.text(1.1, 0, "1", fontsize=17)
plt.text(-0.9, 0, "-1", fontsize=17)
plt.text(0.1, 0.1, "$e^{i\\pi} = -1$", fontsize=17, color = 'r')

# draw lines
plt.axhline(color = "black", linewidth = 0.5)
plt.axvline(color = "black", linewidth = 0.5)

plt.legend(loc = "upper right")
plt.show()
```
