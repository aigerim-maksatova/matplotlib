```python
data = {"Subject" : ["Math", "Art", "Science", "English", "Music", "Computers", "History", "French", "Social Studies", "Geography"],
       ("Elementary, %", "Girls") : [ 14.6, 26.7, 7.1, 8.1, 7.2, 2.2, 2.7, 3.8, 1.4, 1.1], 
       ("Elementary, %", "Boys") : [ 19.8, 7.2, 10.3, 2.2, 3.2, 7.9, 2.8, 1.2, 1.0, 0.8], 
       ("Elementary, %", "All Students") : [17.1, 17.3, 8.6, 5.2, 5.3, 4.9, 2.7, 2.5, 1.2, 1.0],
       ("Secondary, %", "Girls") : [15.2, 16.1, 11.5, 11.8, 6.8, 1.7, 3.3, 2.4, 6.3, 1.3],
       ("Secondary, %", "Boys") : [18.1, 3.6, 11.9, 5.6, 4.6, 6.6, 4.2, 0.8, 2.7, 1.4], 
       ("Secondary, %", "All Students") : [16.6, 10.0, 11.7, 8.8, 5.7, 4.1, 3.7, 1.6, 4.5, 1.3] }
df = pd.DataFrame(data)
df.set_index("Subject", inplace = True)
# df.columns = pd.MultiIndex.from_tuples(df)
# df.set_index("Subject", inplace = True)

# df.columns = pd.MultiIndex.from_tuples([("Elementary, %", "Girls"), ("Elementary, %", "Boys"), ("Elementary, %", "All Students")])
# print(df)
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))  # 1 row, 2 columns

boys = df.iloc[:, 1]
ax1.pie(boys, labels=df.index, autopct='%1.0f%%', pctdistance = 0.9, startangle=90)
ax1.set_title("Elementary, Boys", fontsize = 25, fontweight = "bold", pad = 20)

girls = df.iloc[:, 0]
ax2.pie(girls, labels=df.index, autopct='%1.0f%%',  pctdistance = 0.9, startangle=90)
ax2.set_title("Elementary, Girls", fontsize = 25, fontweight = "bold", pad = 20)

plt.tight_layout()

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6)) 

boys = df.iloc[:, 4]
ax1.pie(boys, labels=df.index, autopct='%1.0f%%', pctdistance = 0.9, startangle=90)
ax1.set_title("Secondary, Boys", fontsize = 25, fontweight = "bold", pad = 20)

girls = df.iloc[:, 3]
ax2.pie(girls, labels=df.index, autopct='%1.0f%%',  pctdistance = 0.9, startangle=90)
ax2.set_title("Secondary, Girls", fontsize = 25, fontweight = "bold", pad = 20)

plt.tight_layout()

plt.show()
```
