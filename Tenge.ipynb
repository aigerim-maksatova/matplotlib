```python
plt.figure(figsize = (12, 6))
plt.plot(snz93_20["Годы"], snz93_20["Доллары"], label="in dollars", marker="o") 
plt.plot(snz93_20["Годы"], snz93_20["Тенге"], label="in tenge", marker="s", color="green")

plt.xlabel("Year")
plt.ylabel("Salary")
plt.title("Среднемесячная Зарплата 1993-2000")
plt.legend()
plt.grid(linewidth=0.4, linestyle="--")
plt.show()
```

```python
salaries = snz93_20["Тенге"].tolist()
years = snz93_20["Годы"].tolist()
```

```python
import numpy as np
years = np.arange(1993, 2021)

plt.figure(figsize=(10, 8))
plt.plot(years, salaries, label= "Зарплата", marker='o')

# the natural logarithm of the salaries:
log_salaries = np.log(salaries)

# log(salary) = log(A) + B * years      LINEAR FORM 
coefficients = np.polyfit(range(len(years)), log_salaries, 1)

# y = A * e^(B * t)                   EXPONENTIAL
A = np.exp(coefficients[1]) 
B = coefficients[0]          
fitted_salaries = A * np.exp(B * range(len(years)))


plt.plot(years, fitted_salaries)
plt.xlabel("Годы")
plt.ylabel("Зарплата в Тенге")
plt.legend()
plt.grid(linewidth = 0.4)
plt.show()
```

```python
import numpy as np
ratios = [salary[i+1] / salary[i] for i in range(len(salary)-1)]
print("Ratios between consecutive values:", ratios)
plt.figure(figsize = (10, 3))
plt.plot(ratios)
plt.show()
# If the ratios are approximately constant, it indicates exponential growth.
```
